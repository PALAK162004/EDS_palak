```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df=pd.read_csv("datasetproject.csv")

X = df[['Mat', '100']]

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_

df['Cluster'] = labels

plt.scatter(df['Player'], df['100'], c=labels)
plt.xlabel('Player')
plt.ylabel('100')
plt.title('Clustering of Players')
plt.show()

print(df[['Player', 'Mat', '100', 'Cluster']])
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df=pd.read_csv("datasetproject.csv")

X = df[['Ave']]

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_

df['Cluster'] = labels

plt.scatter(df['Player'], df['Ave'], c=labels)
plt.xlabel('PLAYER')
plt.ylabel('AVERAGE')
plt.title('Clustering of Players')
plt.show()

print(df[['Player', 'Runs', '100', 'Cluster']])
```
